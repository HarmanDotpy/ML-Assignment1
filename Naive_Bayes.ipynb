```python
import pandas as pd
import numpy as np
from math import sqrt
from math import pi
from math import exp

df = pd.read_csv('health_data.csv')

rows = df.shape[0]

#randomize indces, take the first 80% of the indices and last 20 percent as test
indices = np.random.permutation(df.shape[0])
train_ind, test_ind = indices[:int(0.8*df.shape[0])], indices[int(0.8*df.shape[0]):]
# print(len(train_ind), len(test_ind))

#select the data corresponding to the train and test indices and save into 2 dataframes. Reset index afterwards
train_df, test_df = df.loc[train_ind, :], df.loc[test_ind, :]
train_df, test_df = train_df.reset_index(), test_df.reset_index()

#print('Train data is : \n', train_df.head(), ' \nnumber of rows = {}'.format(train_df.shape[0]))
#print('Test data is : \n', test_df.head(), ' \nnumber of rows = {}'.format(test_df.shape[0]))


#split input by output value

#
#y_split = df.groupby(df.category)

y_split_train = train_df
y_split_test = test_df

#for category, y_table in y_split:
#    train_size = int(0.7*y_table.shape[0])
#    test_size = y_table.shape[0] - train_size
#    y_split_train.append(y_table.head(train_size))
#    y_split_test.append(y_table.tail(test_size))

priors = dict()

y_split_train = y_split_train.groupby(y_split_train.category)
y_split_test = y_split_test.groupby(y_split_test.category)

#find mean and std-dev for each class (output value) to compute class probabilities
for category, y_table in y_split_train:
    priors[category] = y_table.shape[0]/rows

#find mean and std-dev for each feature conditional on output value
conditional_priors = dict()

for category, y_table in y_split_train:
    for (colname, colitems) in y_table.iteritems():
        conditional_priors[colname] = [(colitems.mean(), colitems.std())] 
        
print(conditional_priors)

#use the mean and std-dev for each feature to calculate the prior and use prior probabilities to compare     

def gaussian(x, mean_std_tuple):
    mean = mean_std_tuple[0][0]
    std = mean_std_tuple[0][1]
    return (1/(sqrt(2*pi)*std))*exp(-((x-mean)**2/(2*std**2)))



for name, group in y_split_test:
    for i in range(group.shape[0]):
        post_num = 1
        j = 1
        for col in group.columns[1:-1]:
            post_num = post_num*gaussian(group.iat[i,j], conditional_priors[col])
            j = j+1
        print(post_num)
                                   

    
```
