```python
#undifferentiated cost function: 1/2*X.shape[0]*(h(x)-y)'(h(x) - y)
#update rule: w_i = w_i - a*[1/X.shape[0]*sum(h(x_i) - y_i)*x_i - 2*L*w_i]

def gd_sq_reg(W, X, Y,num_iter,alpha,lam):
    m = X.shape[1]
    for i in range(0,num_iter):
        h_x = np.matmul(np.transpose(W),X)
        print(X.shape)
        print(h_x.shape)
        #print(h_x)
        #print(Y)
        loss = h_x - Y
        print(loss.shape)
        cost = np.linalg.norm(loss)/m
        part_diff = np.matmul(loss,np.transpose(X))
        part_diff = np.transpose(part_diff)
        gradient = part_diff + 2*lam*W
        print(gradient)
        W = W - alpha*gradient
    return W

X = x_train.to_numpy()
X = np.transpose(X)
print(X.shape)
W = np.random.random((X.shape[0],1))
print(W.shape)
Y = y_train.to_numpy()
Y = np.transpose(Y)
print(Y.shape)
num_iter = 10
alpha = 0.01
lam = 0.01

W = gd_sq_reg(W,X,Y,num_iter,alpha,lam)

print(W)
```
